```python
import pandas as pd
import ast
```

```python
filas = []
with open('C:/Users/I5/Documents/Kevin/Henry/Proyectos Individuales II/Proyecto Integrador I/Data/steam_games.json', 'r') as f:
    for i in f.readlines():
        filas.append(ast.literal_eval(i))
```

```python
data = pd.DataFrame(filas)
data
```

```python
data.info()
```

# Se créa una columna llamada año para tener mayor prolijidad en los códigos de los endpoints

Verificar y convertir tipos de datos

```python
data['release_date'] = data['release_date'].astype(str)
```

Identificar y manejar valores no válidos

```python
invalid_values = data[~data['release_date'].str.match(r'\d{4}-\d{2}-\d{2}')]
data = data[~data['release_date'].isin(invalid_values['release_date'])]
```

Convertir la columna 'release_date' en un objeto de fecha y hora

```python
data['release_date'] = pd.to_datetime(data['release_date'])
data['año'] = data['release_date'].dt.year
```

Se transforma la columna año a tipo string para una mejor ejecución en los codigos

```python
data['año'] = data['año'].astype(str)
```
